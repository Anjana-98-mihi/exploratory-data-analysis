## DATA CLEANING

STEP 5: FINDING OUT THE NULLS

```python
print("\n Missing Values per Column:\n", df.isnull().sum())

```

FILLING THE MISSING VALUES

```python
import pandas as pd
df = pd.read_csv('automobileEDA.csv')
df = df.replace("?", pd.NA)
df = df.fillna(df.mean(numeric_only=True))
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("✅ Missing values handled successfully!")


```

STEP 6: FINDING THE DUPLICATES

```python

duplicates = df.duplicated().sum()
print("Duplicates rows:", duplicates)
df.drop_duplicates(inplace=True)
```

STEP 7: FIXING THE TYPES

```python
for col in df.columns:
    if df[col].dtypes == 'object':
        try:
            df[col] = df[col].astype(float)
        except:
            pass

```

STEP 8: N UNIQUE

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

```

## OUTLIERS HANDLING

STEP 14: OUTLIERS DETECTION

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"{col}: {len(outliers)} outliers detected")
```

INTERACTIVE SCATTER

```python
if len(numeric_cols) >= 2:
    fig = px.scatter(df, x=numeric_cols[0], y=numeric_cols[1],
                     title=f"Interactive Scatter: {numeric_cols[0]} vs {numeric_cols[1]}",
                     color=df[numeric_cols[1]])
    fig.show()
```
